```python
# All Different types of graph implementations
class UDUWGraphMatrix:
	"""
	Implementation of an Unweighted, Undirected Graph using a matrix
	(Assuming no self loops, and 0 indexing) 
	"""
	def __init__(self, vertices, graph=None):
		self.v = vertices

		if(graph == None):
			self.graph = [[0 for i in range(self.v)] for j in range(self.v)]
		else:
			self.graph = graph

	def add_edge(self, a, b):
		#method to add an edge
		self.graph[a][b] = 1
		self.graph[b][a] = 1

	def __str__(self):
		return self.graph

	def degree(self, n):
		# returns the degree of node n
		degree = 0
		for i in range(self.v):
			if(self.graph[n][i] == 1):
				degree += 1
		return degree


class UDWGraphMatrix:
	"""
	Implementation of an Weighted, Undirected Graph using a matrix
	(Assuming no self loops, and 0 indexing)
	(Valid only for positive weights)
	"""
	def __init__(self, vertices, graph=None):
		self.v = vertices

		if(graph == None):
			self.graph = [[0 for i in range(self.v)] for j in range(self.v)]
		else:
			self.graph = graph

	def add_edge(self,a, b, w):
		# w is the weight of the edge
		#method to add an edge
		self.graph[a][b] = w
		self.graph[b][a] = w

	# def __str__(self):
	# 	return self.graph

	def degree(self, n):
		# returns the degree of node n
		degree = 0
		for i in range(self.v):
			if(self.graph[n][i] > 0):
				degree += 1
		return degree



class DUWGraphMatrix:
	"""
	Implementation of an Weighted, Undirected Graph using a matrix
	(Assuming no self loops, and 0 indexing) 
	"""
	def __init__(self, vertices, graph=None):
		self.v = vertices

		if(graph == None):
			self.graph = [[0 for i in range(self.v)] for j in range(self.v)]
		else:
			self.graph = graph

	def add_edge(self, a, b):
		# edge directed FROM a TO b
		# w is the weight of the edge
		#method to add an edge
		self.graph[a][b] = 1

	def __str__(self):
		return self.graph

	def degree(self, n):
		# returns the indegree and outdegree of node n
		# returns a tuple where first element is indegree, second element is outdegree
		indegree = 0
		outdegree = 0
		for i in range(self.v):
			if(self.graph[n][i] == 1):
				outdegree += 1
			if(self.graph[i][n] == 1):
				indegree += 1
		return (indegree, outdegree)


class DWGraphMatrix:
	"""
	Implementation of an Weighted, Undirected Graph using a matrix
	(Assuming no self loops, and 0 indexing) 
	(Valid only for positive weights)
	"""
	def __init__(self, vertices, graph=None):
		self.v = vertices

		if(graph == None):
			self.graph = [[0 for i in range(self.v)] for j in range(self.v)]
		else:
			self.graph = graph

	def add_edge(self, a, b, w):
		# edge directed FROM a TO b
		# w is the weight of the edge
		#method to add an edge
		self.graph[a][b] = w

	def __str__(self):
		return self.graph

	def degree(self, n):
		# returns the indegree and outdegree of node n
		# returns a tuple where first element is indegree, second element is outdegree
		indegree = 0
		outdegree = 0
		for i in range(self.v):
			if(self.graph[n][i] >= 0):
				outdegree += 1
			if(self.graph[i][n] >= 0):
				indegree += 1
		return (indegree, outdegree)
```

```python

def clustering_coeff(graph, n):
	"""
	funtion that returns the clustering coefficient of a node in a graph(matrix)
	"""
	neighbours = []
	for i in range(len(graph)):
		if(graph[n][i] > 0):
			neighbours.append(i)
	K = len(neighbours)
	if(K<=1):
		return 0
	N = 0
	for i in range(K):
		for j in range(i,K):
			if(graph[neighbours[i]][neighbours[j]]>0):
				N+=1

	return (2*N)/((K-1)*K)
def average_clustering_coeff(graph):
  clusters = []
  for i in range(len(graph)):
    clusters.append(clustering_coeff(graph, i))
  return(sum(clusters)/len(clusters))


#demo using Les Misrables characters
node = int(input("enter node no "))
print("Clustering Coeff for node "+str(node)+" is: "+str(clustering_coeff(NG.graph, node)))
print("Average Clustering Coeff is: "+str(average_clustering_coeff(NG.graph)))
```
